```python
def create_student_xml():
    root = ET.Element("students")
    
    # Dữ liệu mẫu
    data = [
        {"id": "SV01", "name": "Nam", "year": "2003", "class": "DS01", "gender": "Nam"},
        {"id": "SV02", "name": "Lan", "year": "2003", "class": "DS01", "gender": "Nữ"}
    ]
    
    for item in data:
        s = ET.SubElement(root, "student")
        for key, val in item.items():
            child = ET.SubElement(s, key)
            child.text = val
            
    tree = ET.ElementTree(root)
    tree.write("students.xml", encoding="utf-8", xml_declaration=True)
    print("Đã tạo file students.xml")

create_student_xml()
```

```python
from xml.dom import minidom

# Giả sử nội dung file sample.xml từ Bài 2.3
xml_content = """
<company>
    <staff id="1">
        <name>GeeksForGeeks Company</name>
        <salary>8.5 LPA</salary>
    </staff>
    <staff id="2">
        <name>Akbhar Khan</name>
        <salary>6.5 LPA</salary>
    </staff>
</company>
"""
with open("sample.xml", "w") as f: f.write(xml_content)

# Bài 2.4 & 2.5
doc = minidom.parse("sample.xml")
staffs = doc.getElementsByTagName("staff")

print(f"Số lượng nhân viên: {len(staffs)}")
for s in staffs:
    name = s.getElementsByTagName("name")[0].firstChild.data
    print(f"Tên thành viên: {name}")
```

```python
import json

# 1. Tạo dữ liệu giả định 
data_json = """
{
    "cong_ty": "Công ty TNHH Đất Việt",
    "dia_chi": "abc Giải Phóng - Hà Nội",
    "nhan_vien": [
        {"ten": "Nguyen Van A", "don_vi": "Đơn vị A1"},
        {"ten": "Tran Thi B", "don_vi": "Đơn vị A1"},
        {"ten": "Le Van C", "don_vi": "Đơn vị A2"},
        {"ten": "Pham Thi D", "don_vi": "Đơn vị A3"},
        {"ten": "Hoang Van E", "don_vi": "Đơn vị A2"},
        {"ten": "Doan Van F", "don_vi": "Đơn vị A4"},
        {"ten": "Vu Thi G", "don_vi": "Đơn vị A1"}
    ]
}
"""

def thong_ke_nhan_vien(json_str):
    # Đọc dữ liệu JSON
    data = json.loads(json_str)
    
    ten_cty = data['cong_ty']
    dia_chi = data['dia_chi']
    danh_sach_nv = data['nhan_vien']
    tong_so_nv = len(danh_sach_nv)
    
    # Dùng dictionary để đếm số nhân viên mỗi đơn vị
    # Key: tên đơn vị, Value: số lượng
    thong_ke = {}
    for nv in danh_sach_nv:
        dv = nv['don_vi']
        thong_ke[dv] = thong_ke.get(dv, 0) + 1
    
    # 2. Hiển thị kết quả theo định dạng trong sách
    print(f"Tên công ty: {ten_cty}")
    print(f"Địa chỉ: {dia_chi}")
    print("-" * 5 + " Thống kê nhân viên theo đơn vị " + "-" * 5)
    
    stt = 1
    # Sắp xếp theo tên đơn vị để hiển thị cho đẹp
    for dv in sorted(thong_ke.keys()):
        so_luong = thong_ke[dv]
        ty_le = (so_luong / tong_so_nv) * 100
        
        print(f"{stt}./ Tên đơn vị: {dv}")
        print(f"   - Số nhân viên: {so_luong}")
        print(f"   - Tỷ lệ: {ty_le:.2f}%")
        stt += 1

# Chạy chương trình
thong_ke_nhan_vien(data_json)
```

```python
import json
import os

def ghi_giao_dich():
    filename = "giao_dich.json"
    ds_giao_dich = []
    
    if os.path.exists(filename):
        with open(filename, 'r') as f: ds_giao_dich = json.load(f)

    while True:
        loai = input("Nhập loại giao dịch (Tiền/Vàng): ")
        so_luong = input("Nhập số lượng: ")
        ds_giao_dich.append({"loai": loai, "so_luong": so_luong})
        
        tiep_tuc = input("Bạn có muốn ghi tiếp không? (1: Có, 0: Không): ")
        if tiep_tuc == "0":
            with open(filename, 'w') as f:
                json.dump(ds_giao_dich, f, indent=4)
            print(f"Đã lưu vào {filename}")
            break

# ghi_giao_dich() 
```
