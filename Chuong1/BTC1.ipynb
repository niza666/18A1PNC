```python
class HinhChuNhat:
    def __init__(self):
        self.chieu_dai = 0
        self.chieu_rong = 0

    def nhap(self):
        self.chieu_dai = float(input("Nhập chiều dài: "))
        self.chieu_rong = float(input("Nhập chiều rộng: "))

    def tinh_chu_vi(self):
        return (self.chieu_dai + self.chieu_rong) * 2

    def tinh_dien_tich(self):
        return self.chieu_dai * self.chieu_rong

    def xuat(self):
        print(f"HCN [{self.chieu_dai} x {self.chieu_rong}] -> Chu vi: {self.tinh_chu_vi()}, Diện tích: {self.tinh_dien_tich()}")

# Chạy thử bài 1
hcn = HinhChuNhat()
hcn.nhap()
hcn.xuat()
```

```python
class ThiSinh:
    def __init__(self):
        self.ho_ten = ""
        self.toan = 0
        self.ly = 0
        self.hoa = 0

    def nhap(self):
        self.ho_ten = input("Nhập họ tên thí sinh: ")
        self.toan = float(input("Điểm Toán: "))
        self.ly = float(input("Điểm Lý: "))
        self.hoa = float(input("Điểm Hóa: "))

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def xuat(self):
        print(f"TS: {self.ho_ten} | Tổng điểm: {self.tong_diem()}")

# Chương trình chính bài 2
danh_sach = []
n = int(input("Nhập số lượng thí sinh: "))
for i in range(n):
    print(f"--- Nhập thí sinh thứ {i+1} ---")
    ts = ThiSinh()
    ts.nhap()
    danh_sach.append(ts)

print("\n--- Danh sách thí sinh trúng tuyển (Tổng > 20) ---")
for ts in danh_sach:
    if ts.tong_diem() > 20:
        ts.xuat()
```

```python
class PhanSo:
    def __init__(self, tu=0, mau=1):
        if mau == 0:
            raise ValueError("Mẫu số phải khác 0")
        self.tu = tu
        self.mau = mau

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        while True:
            self.mau = int(input("Nhập mẫu số (khác 0): "))
            if self.mau != 0: break
            print("Mẫu phải khác 0, vui lòng nhập lại!")

    def xuat(self):
        print(f"Phân số: {self.tu}/{self.mau}")

# Chạy thử
ps = PhanSo()
ps.nhap()
ps.xuat()
```

```python
class Stack:
    def __init__(self, capacity=20):
        self.capacity = capacity
        self.data = []

    def is_empty(self):
        return len(self.data) == 0

    def is_full(self):
        return len(self.data) >= self.capacity

    def push(self, value):
        if not self.is_full():
            self.data.append(value)
        else:
            print("Stack đã đầy!")

    def pop(self):
        if not self.is_empty():
            return self.data.pop()
        print("Stack rỗng!")
        return None

    def count(self): # Bài 5
        return len(self.data)

    def print_stack(self): # Bài 6
        print("Nội dung Stack:", self.data)

# Chạy thử Stack
s = Stack(5)
s.push(10.5)
s.push(20.1)
s.print_stack()
print("Số phần tử:", s.count())
```

```python
class Date:
    def __init__(self, d=1, m=1, y=2024):
        self.d, self.m, self.y = d, m, y

    def display(self):
        print(f"{self.d:02d}/{self.m:02d}/{self.y}")

    def is_leap(self, y):
        return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)

    def next_day(self):
        days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.is_leap(self.y): days_in_month[2] = 29
        
        self.d += 1
        if self.d > days_in_month[self.m]:
            self.d = 1
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1
        return self

# Chạy thử
today = Date(31, 12, 2024)
today.display()
today.next_day().display()
```

```python
class Employee:
    def __init__(self, ho_ten="", ngay_sinh=None, ngay_vao_lam=None):
        self.ho_ten = ho_ten
        self.ngay_sinh = ngay_sinh if ngay_sinh else Date()
        self.ngay_vao_lam = ngay_vao_lam if ngay_vao_lam else Date()

    def nhap(self):
        self.ho_ten = input("Nhập tên nhân viên: ")
        print("Nhập ngày sinh:")
        self.ngay_sinh = Date(int(input("Ngày: ")), int(input("Tháng: ")), int(input("Năm: ")))
        print("Nhập ngày vào làm:")
        self.ngay_vao_lam = Date(int(input("Ngày: ")), int(input("Tháng: ")), int(input("Năm: ")))

    def xuat(self):
        print(f"NV: {self.ho_ten}")
        print("Ngày sinh: ", end=""); self.ngay_sinh.display()
        print("Ngày vào làm: ", end=""); self.ngay_vao_lam.display()

# Chạy thử
nv = Employee()
# nv.nhap() # Bỏ comment để nhập từ bàn phím
nv.xuat()
```

```python
import math

class DaGiac:
    def nhap(self): pass
    def tinh_dien_tich(self): pass

class TamGiac(DaGiac):
    def __init__(self, a=0, b=0, c=0):
        self.a, self.b, self.c = a, b, c
    def tinh_dien_tich(self):
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s-self.a) * (s-self.b) * (s-self.c))

class HinhChuNhat_KeThua(DaGiac):
    def __init__(self, l=0, w=0):
        self.l, self.w = l, w
    def tinh_dien_tich(self):
        return self.l * self.w

class HinhVuong(HinhChuNhat_KeThua):
    def __init__(self, a=0):
        super().__init__(a, a)

# Test
hv = HinhVuong(5)
print("Diện tích hình vuông cạnh 5:", hv.tinh_dien_tich())
```

```python
class Diem:
    def __init__(self, x=0, y=0):
        self.x, self.y = x, y

class Elip(Diem):
    def __init__(self, x, y, r1, r2):
        super().__init__(x, y)
        self.r1, self.r2 = r1, r2
    def tinh_dien_tich(self):
        return math.pi * self.r1 * self.r2

class HinhTron(Elip):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)

# Test
ht = HinhTron(0, 0, 10)
print(f"Diện tích hình tròn bán kính 10: {ht.tinh_dien_tich():.2f}")
```
