```python
import pandas as pd
import re
print("--- 1. Đọc dữ liệu & Thông tin cấu trúc ---")
df_movies = pd.read_csv('movies.csv')

print("5 dòng dữ liệu đầu:")
print(df_movies.head())
print("\nThông tin về kiểu dữ liệu (Data Info):")
df_movies.info()

print("\n" + "="*50 + "\n")

print("--- 2. Xử lý dữ liệu bị thiếu/không hợp lệ ---")
print("Số lượng giá trị Null/NaN trước khi xử lý:")
print(df_movies.isnull().sum())
print("\n")
df_movies_cleaned = df_movies.dropna(subset=['title', 'genres'])

print("Số lượng dòng sau khi loại bỏ các bản ghi có title/genres bị thiếu:")
print(f"Từ {len(df_movies)} dòng ban đầu, còn lại {len(df_movies_cleaned)} dòng.")

print("\n")
def is_valid_title(title):
    return bool(re.search(r'\(\d{4}\)$', title))
df_movies_cleaned['is_valid'] = df_movies_cleaned['title'].apply(is_valid_title)

print("Các dòng có title không hợp lệ:")
print(df_movies_cleaned[df_movies_cleaned['is_valid'] == False])
df_movies_cleaned = df_movies_cleaned[df_movies_cleaned['is_valid'] == True].drop(columns=['is_valid'])

print("\nSố lượng dòng sau khi loại bỏ các title không hợp lệ:")
print(f"Còn lại {len(df_movies_cleaned)} dòng.")

print("\n5 dòng dữ liệu cuối cùng sau khi đã làm sạch:")
print(df_movies_cleaned.tail())

print("\n" + "="*50 + "\n")
```
