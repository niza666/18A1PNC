```python
import pandas as pd 
emails = pd.Series(['buying books at amazon.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com'])
pattern = '[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'

extracted_emails_df = emails.str.extract(f'({pattern})', expand=False).dropna()
extracted_emails_series = extracted_emails_df.rename(index={0: 'rameses@egypt.com', 1: 'matt@t.co', 2: 'narendra@modi.com'})

output = extracted_emails_df.reset_index(drop=True)
output.index = output.index + 1

print("Series emails gốc:")
print(emails)
print("\n" + "-"*20 + "\n")
print("Kết quả trích xuất các email hợp lệ:")
print(output)
```
