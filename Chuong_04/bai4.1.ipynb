```python
import pandas as pd
import numpy as np
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

ser_from_list = pd.Series(mylist)
print("Series từ list:")
print(ser_from_list.head())
print("\n" + "="*30 + "\n")

ser_from_array = pd.Series(myarr)
print("Series từ numpy array:")
print(ser_from_array.head())
print("\n" + "="*30 + "\n")

ser_from_dict = pd.Series(mydict)
print("Series từ dict (keys làm index, values làm data):")
print(ser_from_dict.head())
```
