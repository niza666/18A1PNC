```python
import numpy as np 
import pandas as pd 
np.random.seed(100)
ser = pd.Series(np.random.normal(10, 5, 25))
min_val = ser.min()
print(f"1. Giá trị tối thiểu (min): {min_val:.4f}")

q1 = ser.quantile(0.25)
print(f"2. Phân vị thứ 25 (Q1): {q1:.4f}")

median_val = ser.median()
print(f"3. Trung vị (median): {median_val:.4f}")

q3 = ser.quantile(0.75)
print(f"4. Phân vị thứ 75 (Q3): {q3:.4f}")

max_val = ser.max()
print(f"5. Giá trị tối đa (max): {max_val:.4f}")

print("\n" + "="*30 + "\n")
print("Tất cả các giá trị thống kê (dùng .describe()):")
print(ser.describe())
```
