```python
import pandas as pd
import numpy as np
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser = pd.Series(mydict) 
df = ser.to_frame()
df = df.reset_index().rename(columns={'index': 'Index', 0: 'Value'})

print("Series ban đầu:")
print(ser.head())
print("\n" + "="*30 + "\n")
print("DataFrame sau khi chuyển đổi:")
print(df.head())
```
