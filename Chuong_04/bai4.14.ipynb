```python
import numpy as np 
import pandas as pd 
np.random.seed(100)
truth = pd.Series(range(10))
pred = pd.Series(range(10)) + np.random.random(10) 
mse = np.mean((truth - pred)**2)

print("Truth (Giá trị thực):")
print(truth)
print("\nPred (Giá trị dự đoán):")
print(pred)
print("\n" + "-"*20 + "\n")
print(f"Sai số bình phương trung bình (MSE): {mse:.4f}")
```
