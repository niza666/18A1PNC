```python
import pandas as pd 
ser = pd.Series(['01 Jan 2010', '02-02-2011', '20120303', '2013/04/04', '2014-05-05', '2015-06-06T12:20'])
timeseries = pd.to_datetime(ser,dayfirst=True,errors='coerce')

print("Series gốc:")
print(ser)
print("\n" + "-"*20 + "\n")
print("Series sau khi chuyển đổi sang timeseries (datetime objects):")
print(timeseries)
print(f"\nKiểu dữ liệu của Series mới: {timeseries.dtype}")
```
