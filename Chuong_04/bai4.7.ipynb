```python
import numpy as np 
import pandas as pd 
np.random.seed(100)
ser = pd.Series(np.random.randint(1, 5, size=12))
print("Series ban đầu:")
print(ser)
print("\n" + "-"*20 + "\n")

top2_values = ser.value_counts().nlargest(2).index.tolist()

ser_new = ser.copy()

ser_new[~ser_new.isin(top2_values)] = 'Other'

print(f"Top 2 giá trị tần suất cao nhất: {top2_values}")
print("\nSeries sau khi thay thế:")
print(ser_new)
```
