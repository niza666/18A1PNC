```python
import pandas as pd
data = {
    'country': [
        'Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola', 
        'Antigua & Barbuda', 'Argentina', 'Armenia', 'Australia', 'Austria'
    ],
    'beer_servings': [
        0, 89, 25, 245, 217, 102, 29, 21, 261, 279
    ],
    'spirit_servings': [
        0, 132, 0, 138, 57, 128, 45, 179, 72, 75
    ],
    'wine_servings': [
        0, 54, 14, 312, 45, 56, 117, 11, 212, 191
    ],
    'total_litres_of_pure_alcohol': [
        0.0, 4.9, 0.7, 12.4, 5.9, 5.6, 8.3, 3.8, 10.4, 9.7
    ],
    'continent': [
        'Asia', 'Europe', 'Africa', 'Europe', 'Africa', 
        'North America', 'South America', 'Europe', 'Oceania', 'Europe'
    ]
}

drinks_df = pd.DataFrame(data)

drinks_df.to_csv('drinks.csv', index=False)

print("✅ Đã tạo file 'drinks.csv' thành công trong thư mục hiện tại.")
drink = pd.read_csv('drinks.csv', index_col=0)

print("\n--- Bài 4.21. Xử Lý Dataset drink.csv ---")
print(f"\nKiểu dữ liệu (type) của drink: {type(drink)}")
print(f"Kích thước (shape) của drink: {drink.shape}")
print("\nTên các cột (columns) của drink:")
print(drink.columns.tolist())
print("\n5 dòng đầu (head):")
print(drink.head())
print("\n5 dòng cuối (tail):")
print(drink.tail())
avg_beer_by_continent = drink.groupby('continent')['beer_servings'].mean()
print("\n2. Số lượng bia tiêu thụ trung bình ở mỗi châu lục:")
print(avg_beer_by_continent)

wine_stats_by_continent = drink.groupby('continent')['wine_servings'].describe()
print("\n3. Thông tin thống kê rượu vang (wine_servings) ở mỗi châu lục:")
print(wine_stats_by_continent)

avg_all_drinks_by_continent = drink.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings']].mean()
print("\n4. Số lượng bia và rượu tiêu thụ trung bình ở mỗi châu lục:")
print(avg_all_drinks_by_continent)
median_all_drinks_by_continent = drink.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings']].median()
print("\n5. Giá trị trung vị (median) của bia và rượu ở mỗi châu lục:")
print(median_all_drinks_by_continent)

spirit_metrics = drink.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min'])
print("\n6. Số lượng rượu mạnh (spirit_servings) trung bình, lớn nhất, nhỏ nhất ở mỗi châu lục:")
print(spirit_metrics)

drink_sorted_by_beer = drink.sort_values(by='beer_servings', ascending=True)

print("\n7. 5 quốc gia có lượng tiêu thụ bia ít nhất:")
print(drink_sorted_by_beer.head(5)['beer_servings'])

print("\n5 quốc gia có lượng tiêu thụ bia nhiều nhất:")
print(drink_sorted_by_beer.tail(5)['beer_servings'])
```
