```python
import numpy as np 
import pandas as pd 
np.random.seed(100)
ser = pd.Series(np.random.randint(1, 7, 7)) 
positions = ser[ser % 3 == 0].index.tolist()

print("Series gốc:")
print(ser)
print("\n" + "-"*20 + "\n")
print(f"Các số là bội của 3: \n{ser[ser % 3 == 0]}")
print(f"Vị trí (chỉ mục) của các số là bội của 3: {positions}")
```
