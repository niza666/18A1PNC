```python
import pandas as pd
import numpy as np
np.random.seed(100)
fruit = pd.Series(np.random.choice(['apple', 'banana', 'carrot'], 10))
weights = pd.Series(np.linspace(1, 10, 10))

print("Series fruit:")
print(fruit)
print("\nSeries weights:")
print(weights)
print("\n" + "="*40 + "\n")

df = pd.DataFrame({'Fruit': fruit, 'Weight': weights})

average_weight_by_fruit = df.groupby('Fruit')['Weight'].mean()

print("Kết quả: Khối lượng trung bình của từng loại quả (Average Weight):")
print(average_weight_by_fruit)
print("\n" + "="*40 + "\n")
weighted_sum_by_fruit = df.groupby('Fruit')['Weight'].sum()
print("Kiểm tra: Tổng khối lượng của từng loại quả (Weighted Sum):")
print(weighted_sum_by_fruit)
```
