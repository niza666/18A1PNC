```python
import numpy as np
import pandas as pd
ser1 = pd.Series(range(5))
ser2 = pd.Series(list('abcde'))

vertical_concat = pd.concat([ser1, ser2], axis=0)
print("1. Xếp chồng theo chiều dọc (Series):")
print(vertical_concat)
print("\n" + "-"*20 + "\n")

horizontal_concat = pd.concat([ser1, ser2], axis=1)
horizontal_concat.columns = ['Numbers', 'Characters']
print("2. Xếp chồng theo chiều ngang (DataFrame):")
print(horizontal_concat)
```
