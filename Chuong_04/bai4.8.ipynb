```python
import numpy as np 
import pandas as pd
np.random.seed(100)
ser = pd.Series(np.random.random(20)) 
deciles_num = pd.qcut(ser, q=10, labels=False)

decile_labels = [f'P{i*10}' for i in range(1, 11)]

ser_deciles = pd.qcut(ser, q=10, labels=decile_labels)

print("Series gốc (5 phần tử đầu):")
print(ser.head())
print("\n" + "-"*20 + "\n")
print("Series sau khi thay thế bằng tên phân vị (5 phần tử đầu):")
print(ser_deciles.head())

print("\nSố lượng phần tử trong mỗi phân vị:")
print(ser_deciles.value_counts())
```
