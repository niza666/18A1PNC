```python
import numpy as np
import pandas as pd 
ser1 = pd.Series([10, 9, 6, 5, 3, 1, 12, 8, 13])
ser2 = pd.Series([1, 3, 10, 13])

positions = ser1[ser1.isin(ser2)].index.tolist()

print("Series ser1:")
print(ser1.values)
print("Series ser2:")
print(ser2.values)
print("\n" + "-"*20 + "\n")
print(f"Các mục của ser2 có mặt trong ser1: {ser1[ser1.isin(ser2)].values}")
print(f"Vị trí (chỉ mục) của chúng trong ser1: {positions}")
```
