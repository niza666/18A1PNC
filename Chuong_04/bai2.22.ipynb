```python
import pandas as pd
import numpy as np
print("--- 1. Đọc file CSV và xem thông tin cơ bản ---")

# a) Đọc file stocks1.csv
stocks1 = pd.read_csv('stocks1.csv')
print("\n--- a) stocks1.csv ---")
print("Hiển thị 5 dòng dữ liệu đầu:")
print(stocks1.head())
print("\nThông tin về kiểu dữ liệu (dtype) của các cột:")
print(stocks1.dtypes)
print("\nThông tin (info) đầy đủ:")
stocks1.info()

# b) Đọc file stocks2.csv
stocks2 = pd.read_csv('stocks2.csv')
print("\n--- b) stocks2.csv ---")
print("Hiển thị 5 dòng dữ liệu đầu và cuối:")
print(stocks2.head())
print(stocks2.tail())
print("\nThông tin về kiểu dữ liệu (dtype) của các cột:")
print(stocks2.dtypes)
print("\nThông tin (info) đầy đủ:")
stocks2.info()

# c) Đọc file companies.csv
companies = pd.read_csv('companies.csv')
print("\n--- c) companies.csv ---")
print("Thông tin về kiểu dữ liệu (dtype) của các cột:")
print(companies.dtypes)
print("\nThông tin (info) đầy đủ:")
companies.info()

print("\n" + "="*50 + "\n")

print("--- 2. Xử lý giá trị Null trong stocks1 ---")

print("Số lượng giá trị Null trong stocks1 trước khi xử lý:")
print(stocks1.isnull().sum())
stocks1['high'] = stocks1['high'].fillna(stocks1['high'].max())
stocks1['low'] = stocks1['low'].fillna(stocks1['low'].min())

print("\nSố lượng giá trị Null trong stocks1 sau khi xử lý:")
print(stocks1.isnull().sum())
print("\n" + "="*50 + "\n")

print("--- 3. Gộp dataframe stocks1 và stocks2 ---")

stocks = pd.concat([stocks1, stocks2], ignore_index=True)

print("Kích thước dataframe stocks sau khi gộp:", stocks.shape)
print("Hiển thị 15 dòng dữ liệu cuối của stocks:")
print(stocks.tail(15))
print("\n" + "="*50 + "\n")

print("--- 4. Gộp dataframe stocks và companies ---")

stocks_companies = pd.merge(stocks, companies, on='stock', how='left')

print("Kích thước dataframe stocks_companies sau khi gộp:", stocks_companies.shape)
print("Hiển thị 5 dòng dữ liệu đầu của stocks_companies:")
print(stocks_companies.head())
print("\n" + "="*50 + "\n")

print("--- 5. Tính trung bình theo mỗi công ty ---")
company_averages = stocks_companies.groupby('name')[['open', 'high', 'low', 'close', 'volume']].mean()

print("Giá trị trung bình của các cột theo từng công ty:")
print(company_averages)
print("\n" + "="*50 + "\n")

print("--- 6. Thống kê giá đóng cửa (close)---")
close_stats = stocks_companies.groupby('stock')['close'].agg(['mean', 'max', 'min'])

print("Giá đóng cửa (close) trung bình, lớn nhất và nhỏ nhất của mỗi mã chứng khoán:")
print(close_stats)
print("\n" + "="*50 + "\n")

print("--- 7. Chuyển đổi cột 'date' sang định dạng DateTime ---")
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])

print("Thông tin về kiểu dữ liệu (dtype) sau khi thêm cột 'parsed_time':")
print(stocks_companies.dtypes)
print("\nHiển thị 5 dòng dữ liệu đầu với cột 'parsed_time':")
print(stocks_companies[['date', 'parsed_time']].head())
print("\n" + "="*50 + "\n")

print("--- 8. Thêm cột 'result' ('up'/'down') ---")
stocks_companies['result'] = np.where(stocks_companies['close'] > stocks_companies['open'], 'up', 'down')

print("Hiển thị 10 dòng dữ liệu cuối với cột 'result':")
print(stocks_companies[['open', 'close', 'result']].tail(10))

print("\nSố lượng ngày 'up' và 'down':")
print(stocks_companies['result'].value_counts())
print("\n" + "="*50 + "\n")

```
