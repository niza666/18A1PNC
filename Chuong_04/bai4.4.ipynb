```python
import pandas as pd
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
items_in_ser2 = ser1[ser1.isin(ser2)]
print("1. Các mục của ser1 có mặt trong ser2:")
print(items_in_ser2)
items_not_in_ser2=ser1[~ser1.isin(ser2)]
print("\n" + "-"*20 + "\n")
print("2. Các mục trong ser1 KHÔNG có mặt trong ser2:")
print(items_not_in_ser2)
```
