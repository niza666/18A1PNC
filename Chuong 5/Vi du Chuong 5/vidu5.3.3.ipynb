#### Ví dụ 5.3.3.1
* Giả sử có CSDL mydatabase.db. Vẽ biểu đồ biểu diễn tuổi của các nhân viên trong CSDL.

```python
import sqlite3
import matplotlib.pyplot as plt

# Kết nối đến cơ sở dữ liệu
conn = sqlite3.connect('mydatabase.db')
cursor = conn.cursor()

# Truy vấn dữ liệu từ cơ sở dữ liệu
cursor.execute('SELECT "Tên SV", "Tuổi" FROM users')
rows = cursor.fetchall()     #Biến rows là một list lưu kết quả của câu lệnh SQL, 
                             #hàm fetchall() trả về một danh sách các bản ghi từ câu lệnh SQL
                             #Mỗi bản ghi trong rows tương ứng với một hàng trong bảng dữ liệu 

# Tách dữ liệu thành hai danh sách riêng biệt cho tên và tuổi
Tên = [row[0] for row in rows]      #Truy cập dữ liệu trong mỗi bản ghi bằng chỉ mục
Tuổi = [row[1] for row in rows]

# Đóng kết nối
conn.close()
```

```python
# Vẽ biểu đồ cột
plt.bar(Tên, Tuổi, width=0.2)
plt.xlabel('Họ tên')
plt.ylabel('Độ tuổi')
plt.title('Biểu đồ tuổi của nhân viên')
plt.show()
```

Để tô màu cho các cột khác nhau trong biểu đồ cột,  
có thể sử dụng thêm tham số color trong hàm plt.bar()

```python
import sqlite3
import matplotlib.pyplot as plt

# Kết nối đến cơ sở dữ liệu
conn = sqlite3.connect('mydatabase.db')
cursor = conn.cursor()

# Truy vấn dữ liệu từ cơ sở dữ liệu
cursor.execute('SELECT "Tên SV", "Tuổi" FROM users')
rows = cursor.fetchall()

# Tách dữ liệu thành hai danh sách riêng biệt cho tên và tuổi
Tên = [row[0] for row in rows]
Tuổi = [row[1] for row in rows]

# Đóng kết nối
conn.close()

# Màu cho từng cột, ví dụ: mỗi cột có một màu khác nhau
colors = ['blue', 'green', 'red', 'purple', 'orange']

# Vẽ biểu đồ cột với màu sắc được chọn
plt.bar(Tên, Tuổi, width=0.2, color=colors)
plt.xlabel('Họ tên')
plt.ylabel('Độ tuổi')
plt.title('Biểu đồ tuổi của nhân viên')
plt.show()
```

#### Ví dụ 5.3.3.2
Giả sử có số liệu cơ sở dữ liệu SQLite ‘sales.db’ về doanh thu hàng năm của một công ty từ năm 2012 đến năm 2022. Hãy vẽ biểu đồ dạng line Chart thể hiện doanh thu theo các năm của công ty đó.


```python
import sqlite3
import matplotlib.pyplot as plt

# Kết nối đến cơ sở dữ liệu
conn = sqlite3.connect('sales.db')
cursor = conn.cursor()

# Truy vấn dữ liệu từ cơ sở dữ liệu
cursor.execute("SELECT year, revenue FROM sales")
rows = cursor.fetchall()

# Tách dữ liệu thành hai danh sách riêng biệt cho năm và doanh thu
years = [row[0] for row in rows]
revenues = [row[1] for row in rows]

# Đóng kết nối
conn.close()

# Vẽ biểu đồ Line Chart
plt.plot(years, revenues, marker='x', linestyle='-')
plt.xlabel('Các năm')
plt.ylabel('Doanh thu (đơn vị: tỷ đồng)')
plt.title('Biểu đồ biểu diễn doanh thu theo năm')

plt.legend(['Doanh thu theo năm'])  # Thêm chú thích vào biểu đồ
plt.grid(True)  # Thêm lưới trên biểu đồ
plt.show()
```

#### Ví dụ 5.3.3.3
* Vẽ biểu đồ mức chi tiêu hàng tháng của một hộ gia đình.

```python
import sqlite3
import matplotlib.pyplot as plt

# Kết nối đến cơ sở dữ liệu hoặc tạo CSDL mới nếu nó chưa tồn tại
conn = sqlite3.connect('ct_giadinh.db')
cursor = conn.cursor()     #Tạo đối tượng con trỏ cursor từ đối tượng kết nối conn

# Truy vấn dữ liệu từ bảng 'khoan_chi'
cursor.execute("SELECT * FROM khoan_chi")
rows = cursor.fetchall()

# Lấy thông tin về các tên cột
column_names = [d[0] for d in cursor.description][1:]

# Tách dữ liệu thành hai danh sách riêng biệt cho loại chi tiêu và tổng số tiền
loai_chi = column_names
tien_chi = rows[0][1:]

# Đóng kết nối
conn.close()

# Vẽ biểu đồ Pie Chart
plt.pie(tien_chi, labels=loai_chi, autopct='%1.2f%%')
plt.title('Biểu đồ tỉ lệ mức chi tiêu một hộ gia đình')
plt.axis('equal')  # Để đảm bảo hình tròn
plt.show()
```
